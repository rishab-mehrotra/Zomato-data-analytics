PROPORTION OF RESTAURANTS OFFERING ONINE ORDERS OR TABLE BOOKING


6. GROUPBY ARA OR RESTAURENT TYPE AND CALCULATE THE PROPORTION OF YES IN ONLINE ORDER OR TABLE BOOKING . 
PREVALENT ARE ONLINE ORDERS AND TABLE BOOKINGS IN DIFFERENT AREAS ?

```python
online_table_proportion=data.groupby('area')[['online_order','table booking']].apply(lambda x:(x=='Yes').mean())
online_table_proportion
```

```python
online_table_proportion.plot(kind='barh',title='proportion of restaurents offering online orders or table booking',ylabel='area',xlabel='proportion')
plt.show()
```

ANALYSZE BINARY COLUMNS (ONLINE_ORDER AND TABLE BOOKING)

MELT THESE COLUMNS INTO A LING FORMAT WITH RESTAURANT NAME AR AREA TO SEE TRENDS IN ONLINE ORDERING AND TABLE BOOKING.


INSIGHT :----- WHICH AREAS OR RESTAURANT TYPES ARE MORE TECH-FRIENDLY

```python
binary_melt=pd.melt(data,id_vars=['restaurant name','area'],value_vars=['online_order','table booking'])
binary_melt
```

```python
binary_trend=binary_melt.groupby(['area','variable','value']).size().unstack().fillna(0)[:20]
binary_trend
```

```python
binary_trend.plot(kind='bar',stacked=True,title='trend in online orders and table booking by area')
plt.show()
```
